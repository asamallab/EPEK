```python
#import neccessary packages
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read the data from the excel file
goenrich = pd.read_excel('GO_enrichment.xlsx')
print(goenrich.shape)
goenrich
```

```python
# Creating cutom color pallette for the three ontologies
custompallette = []
for i in range(goenrich.shape[0]):
    if goenrich['OntologyType'][i] == 'Biological Process':
        custompallette.append('#72D6BF')
    elif goenrich['OntologyType'][i] == 'Molecular Function':
        custompallette.append('#FFC18E')
    elif goenrich['OntologyType'][i] == 'Cellular Component':
        custompallette.append('#A2D2FF')

# Plotting the horizontal barplot with the custom color for bars
plt.figure(figsize = (3,6.5))
sns.barplot(data=goenrich,x='NofGenes',y='OntologyTerm',palette=custompallette)
plt.xlabel('Number of Genes')
plt.ylabel('GO term')
plt.title('GO Term Enrichment')
## Save figure to pdf
#plt.savefig('GO_enrichment.pdf',bbox_inches = 'tight')
plt.grid(axis='x')
plt.show()
```
